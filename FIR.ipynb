```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\DELL\Downloads\simplified_case_data_1000_rows.csv")
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(df['Case Description'],df['Applicable Sections'],test_size = 0.2,random_state = 42)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(max_features=5000)

X_train_vec = tfidf.fit_transform(X_train)

X_test_vec = tfidf.transform(X_test)
```

```python
X_train_vec.shape,X_test_vec.shape,y_train.shape,y_test.shape
```

```python
from sklearn.preprocessing import LabelEncoder

encode = LabelEncoder()

y_train_encode = encode.fit_transform(y_train)

y_test_encode = encode.transform(y_test)
```

So,

OnhotEncoder and LabelEncoder both convert catergroical values to numerical but serve differently

oneHotEncoeder takes input as a 2D matrix but labelEncoder takes input as 1D array

whereas using both the methods there are disadvantages

and Onehotencoding is used for dependent data and labelEncoding for independent data

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()

model.fit(X_train_vec,y_train_encode)
```

```python
y_pred = model.predict(X_test_vec)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score,classification_report

accuracy_score(y_pred,y_test_encode)*100
```
